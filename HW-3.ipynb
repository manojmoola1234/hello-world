```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
typesofdata = {
    "Year": [1990, 1995, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009],
    "White non-Hispanic": [169600, 203300, 260500, 271700, 291800, 301200, 317400, 331000, 336500, 338200, 333300, 326400],
    "Black non-Hispanic": [172300, 220600, 256300, 256200, 264900, 271000, 275400, 290500, 295900, 301700, 308000, 300500],
    "Hispanic": [58100, 74400, 94100, 93000, 98000, 106600, 108300, 111900, 119200, 125500, 128500, 124000],
    "Other": [5400, 8800, 10200, 10300, 10800, 12500, 12900, 13000, 13500, 13900, 14000, 14800]
}
df = pd.DataFrame(typesofdata)
sns.set(style="whitegrid")
plt.figure(figsize=(12, 5))
sns.scatterplot(data=df, x='Year', y='White non-Hispanic', label='White non-Hispanic', s=100)
sns.scatterplot(data=df, x='Year', y='Black non-Hispanic', label='Black non-Hispanic', s=100)
sns.scatterplot(data=df, x='Year', y='Hispanic', label='Hispanic', s=100)
sns.scatterplot(data=df, x='Year', y='Other', label='Other', s=100)
plt.title("Scatter Plot of Jail Inmates by Race and Hispanic Origin (1990-2009)")
plt.xlabel("Years")
plt.ylabel("No of Inmates")
plt.legend()
sns.despine()
plt.tight_layout()
plt.show()
```
